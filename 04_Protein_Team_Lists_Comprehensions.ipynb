```python
# --- Application: Calcul pour toute une équipe ---

# 1. Liste des poids (kg)
weights_team = [65, 72, 80, 95, 58]

# 2. Le facteur de protéine (Maintien = 1.8)
factor = 1.8

# 3. La "Magie" de List Comprehension
# On calcule le besoin pour chaque poids 'w' dans la liste
protein_needs = [round(w * factor, 1) for w in weights_team]

# 4. Affichage des résultats
print(f"Poids de l'équipe: {weights_team}")
print(f"Besoins en protéines (g): {protein_needs}")
```
