```python
def smart_protein_calculator():
    """
    Application de calcul de protéines avec gestion d'erreurs 
    et utilisation de dictionnaires (Concepts Data Science).
    """
    print("--- Bienvenue dans le Smart Protein Calculator ---")
    
    try:
        # Saisie des informations utilisateur
        nom = input("Entrez votre nom : ")
        # Gestion d'exception si le poids n'est pas un nombre
        poids = float(input("Entrez votre poids (kg) : ")) 

        # Utilisation d'un Dictionnaire (plus efficace que if/elif)
        # Format : "Clé": (Facteur, "Nom de l'objectif")
        objectifs = {
            "1": (2.2, "Prise de masse"),
            "2": (1.8, "Maintien"),
            "3": (2.4, "Perte de poids")
        }

        while True:
            print("\n1. Prise de masse | 2. Maintien | 3. Perte de poids")
            choix = input("Sélectionnez votre objectif (1/2/3) : ")

            if choix in objectifs:
                facteur, mode = objectifs[choix]
                break
            else:
                print("Choix invalide. Veuillez saisir 1, 2 ou 3.")

        # Calcul et affichage des résultats
        total_proteines = poids * facteur
        print(f"\n--- Résultats pour {nom} ---")
        print(f"Objectif : {mode}")
        print(f"Besoin quotidien : {total_proteines:.1f}g de protéines.")

    except ValueError:
        print("Erreur : Veuillez entrer un poids valide (ex: 70.5).")

# Exécution de la fonction
smart_protein_calculator()
```
