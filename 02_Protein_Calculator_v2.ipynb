```python
print("--- Bienvenue dans le Smart Protein Calculator de Hajar ---")

# Étape 1 : Entrées
name = input("Entrez votre nom : ")
weight = float(input("Entrez votre poids (kg) : "))

# Étape 2 : Choix de l'objectif
while True:
    print("\nChoisissez votre objectif :")
    print("1. Prise de masse")
    print("2. Maintien")
    print("3. Perte de poids")
    
    goal = input("Tapez le numéro de l'objectif (1/2/3) : ")
    
    if goal in ["1", "2", "3"]:
        break  # Sort de la boucle si l'entrée est valide
    else:
        print("Entrée invalide. Veuillez choisir 1, 2 ou 3.")

# Étape 3 : Logique (Conditions if/elif/else)
if goal == "1":
    factor = 2.2  # Protéines pour la prise de masse
    mode = "Prise de masse"
elif goal == "2":
    factor = 1.8  # Maintien
    mode = "Maintien"
elif goal == "3":
    factor = 2.4  # Protéines pour la perte de poids pour préserver les muscles
    mode = "Perte de poids"

# Étape 4 : Résultat
total_protein = weight * factor
print(f"\n--- Résultat pour {name} ---")
print(f"Objectif : {mode}")
print(f"Besoin quotidien : {total_protein:.1f}g de protéines.")
```
