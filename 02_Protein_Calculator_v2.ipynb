```python
# Mini-Project: Protein Calculator v2.0 (Logic Version)
print("--- Marhba bik f Smart Protein Calc d Hajar ---")

# Step 1: Inputs
name = input("Entrez votre nom : ")
weight = float(input("Entrez votre poids (kg) : "))

print("\nKhtari l-hadaf dyalk (Objectif) :")
print("1. Bulk (Ziyadt l-3adlat)")
print("2. Maintenance (3adi)")
print("3. Cut (Naqs l-vazn)")
goal = input("Kteb raqm l-hadaf (1/2/3): ")

# Step 2: Logic (Conditions if/elif/else)
if goal == "1":
    factor = 2.2  # Protein tale3 l-bulk
    mode = "Bulk"
elif goal == "2":
    factor = 1.8  # Maintenance
    mode = "Maintenance"
elif goal == "3":
    factor = 2.4  # Protein tale3 ktar f l-cut bach n-7afdo 3la l-muscles
    mode = "Cut"
else:
    factor = 2.0
    mode = "Standard"

# Step 3: Result
total_protein = weight * factor
print(f"\n--- Natija l-{name} ---")
print(f"Objectif: {mode}")
print(f"Besoin quotidien: {total_protein:.1f}g de protéines.")
```
