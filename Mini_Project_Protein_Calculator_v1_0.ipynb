```python
# Mini-Project: Protein Calculator v1.0
print("--- Bienvenue sur Protein Calc by Hajar ---")

# Step 1: User Input (Inputs )
name = input("Entrez votre nom : ")
weight = float(input("Entrez votre poids (kg) : "))

# Step 2: Calculation
# Standard: 2g de protéines par kg
total_protein = weight * 2

# Step 3: Result (Output )
print(f"\nSalut {name}!")
print(f"Selon votre poids ({weight} kg), vous avez besoin de {total_protein}g de protéines par jour.")
```
